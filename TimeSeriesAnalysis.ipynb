Granger Causality Analysis
- Determines if past values of one time series help predict future values of another

	- Significant p-values (p < 0.05): Suggest that one feature Granger-causes another.
	- Application: Useful for feature selection and understanding temporal dependencies.
	- Stationarity: Granger causality tests assume stationarity - check this
	- Multiple Participants: Perform the test separately for each participant or aggregate appropriately.

```python
from statsmodels.tsa.stattools import adfuller

def check_stationarity(series):
    result = adfuller(series)
    p_value = result[1]
    if p_value < 0.05:
        print(f"{series.name} is stationary (p-value: {p_value:.4f})")
    else:
        print(f"{series.name} is not stationary (p-value: {p_value:.4f})")
    return p_value < 0.05


# Apply to each column
df = X
for column in df.columns:
    # check_stationarity(df[column])
    pass
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
import pandas as pd
import numpy as np
```

Apply granger
- Null hypothesis: Past values of X are not useful in predicting Y
, if p < 0.05 reject this.

```python
# Function to perform Granger causality tests between each feature and the target ('activity')
def grangers_causation_with_activity(data, activity_col='activity', maxlag=5, test='ssr_chi2test'):
    variables = data.columns.tolist()
    variables.remove(activity_col)
    df_pvalues = pd.DataFrame(index=variables, columns=[activity_col])
    
    for predictor_var in variables:
        test_result = grangercausalitytests(
            data[[activity_col, predictor_var]], maxlag=maxlag
        )
        p_values = [round(test_result[i + 1][0][test][1], 4) for i in range(maxlag)]
        min_p_value = np.min(p_values)
        df_pvalues.loc[predictor_var, activity_col] = min_p_value

    return df_pvalues

# Combine features and target into one DataFrame
df = pd.concat([X, y], axis=1)

# Perform Granger causality analysis with respect to 'activity'
causality_matrix = grangers_causation_with_activity(df, activity_col='activity', maxlag=5)
```
